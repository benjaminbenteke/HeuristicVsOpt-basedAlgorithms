```python
import matplotlib.pyplot as plt
import numpy as np

Exs = ['Ex1', 'Ex2', 'Ex3', 'Ex4', 'Ex5', 'Ex6', 'Ex7', 'Ex8', 'Ex9']

# Collect your y-values in a dictionary
# data = {
#     'BFGS': [BFGS[m]["pdp"][-1] for m in Exs],
#     'GSM':  [GSM[m]["pdp"][-1]  for m in Exs],
#     'EIA':  [EIA[m]["pdp"][-1]  for m in Exs],
#     'SGD':  [SGD[m]["pdp"][-1]  for m in Exs]
# }

best_k= {
    'BFGS': [np.argmax(BFGS[m]["pdp"]) for m in Exs],
    'GSM':  [np.argmax(GSM[m]["pdp"])  for m in Exs],
    'EIA':  [np.argmax(EIA[m]["pdp"])  for m in Exs],
    'SGD':  [np.argmax(SGD[m]["pdp"])  for m in Exs]

}
best_k
```

```python
# [BFGS[Exs[m]]["pdp"][best_k["BFGS"][m]] for m in range(len(Exs))]
```

```python

# data = {
#     'BFGS': [BFGS[m]["pdp"][-1] for m in Exs],
#     'GSM':  [GSM[m]["pdp"][-1]  for m in Exs],
#     'EIA':  [EIA[m]["pdp"][-1]  for m in Exs],
#     'SGD':  [SGD[m]["pdp"][-1]  for m in Exs]
# }

data = {
    'BFGS': [BFGS[Exs[m]]["pdp"][best_k["BFGS"][m]] for m in range(len(Exs))],
    'GSM':  [GSM[Exs[m]]["pdp"][best_k["GSM"][m]] for m in range(len(Exs))],
    'EIA':  [EIA[Exs[m]]["pdp"][best_k["EIA"][m]] for m in range(len(Exs))],
    'SGD':  [SGD[Exs[m]]["pdp"][best_k["SGD"][m]] for m in range(len(Exs))]
}

colors = {
    'BFGS': 'magenta',
    'GSM': 'black',
    'EIA': 'green',
    'SGD': 'purple'
}

x = np.arange(len(Exs))  # the label locations
width = 0.2  # width of each bar

fig, ax = plt.subplots(figsize=(10,6))

# Plot each method
for i, (method, values) in enumerate(data.items()):
    ax.bar(x + i*width, values, width, label=method, color=colors[method])

# Labels and title
ax.set_xlabel('Examples')
ax.set_ylabel('Percentage of distinct points (PDP)')
ax.set_xticks(x + width*1.5)  # center the group
ax.set_xticklabels(Exs)
ax.legend()

plt.tight_layout()
plt.savefig("dist_points.png", dpi=300)  # save image
plt.show()
```
